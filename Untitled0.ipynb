```python
# Подключаем библиотеки для анализа данных, работы с таблицами и визуализации.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем датасет в виде pandas DataFrame.
df = pd.read_csv("imdb_top_1000.csv")

# Проверяем размер датасета: количество строк и столбцов.
print("Размер датасета:", df.shape)

# Показываем первые строки таблицы, чтобы убедиться, что данные загружены корректно.
df.head()



# На этом этапе начинается предобработка данных.
# Сначала удаляем строки с пропущенными значениями, чтобы модели могли корректно обучиться.
df = df.dropna()

# В столбце Genre у некоторых фильмов указано несколько жанров через запятую.
# Мы берём только первый жанр, чтобы упростить модель и анализ категориальных признаков.
df['Genre'] = df['Genre'].apply(lambda x: x.split(',')[0].strip())

# Столбец Runtime содержит строки вида "142 min".
# Удаляем текст " min" и превращаем значения в целые числа минут.
df['Runtime'] = df['Runtime'].str.replace(' min','').astype(int)

# В зависимости от версии датасета столбец с количеством голосов может называться по-разному.
# Проверяем несколько вариантов и выбираем правильный.
if 'No_of_Votes' in df.columns:
    df['Votes'] = df['No_of_Votes']
elif 'IMDB_Rating_Count' in df.columns:
    df['Votes'] = df['IMDB_Rating_Count']
else:
    # Если нет ни одного из ожидаемых столбцов — выводим ошибку.
    raise ValueError("В датасете нет столбца с количеством голосов.")



# Переходим к разведочному анализу данных (EDA).
# Строим гистограмму распределения рейтингов IMDb.
plt.figure(figsize=(6,4))
sns.histplot(df['IMDB_Rating'], kde=True)
plt.title("Распределение IMDb рейтингов")
plt.show()

# Анализируем самые популярные жанры в датасете.
plt.figure(figsize=(8,4))
df['Genre'].value_counts().head(10).plot(kind='bar')
plt.title("Топ-10 жанров")
plt.show()

# Строим диаграмму зависимости рейтинга от длительности фильма.
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Runtime', y='IMDB_Rating')
plt.title("Зависимость рейтинга от длительности")
plt.show()



# Начинаем подготовку данных для обучения моделей.
# Подключаем необходимые инструменты из библиотеки scikit-learn.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Формируем матрицу признаков X и целевую переменную y.
# Мы используем жанр, длительность и количество голосов как входные данные.
X = df[['Genre', 'Runtime', 'Votes']]
y = df['IMDB_Rating']

# Указываем, какие признаки являются категориальными, а какие — числовыми.
categorical_features = ['Genre']
numeric_features = ['Runtime', 'Votes']

# Создаём преобразователь, который:
# — применит OneHotEncoding к жанрам
# — передаст числовые признаки без изменений
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', 'passthrough', numeric_features)
    ]
)



# Обучение трёх разных моделей машинного обучения для сравнения качества:
# 1. Linear Regression — классическая линейная модель.
# 2. Ridge Regression — линейная модель с L2-регуляризацией, защищает от переобучения.
# 3. RandomForest — ансамбль деревьев решений, хорошо работает с разными типами данных.

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor

# Каждая модель помещена в Pipeline, где сначала выполняется предобработка данных,
# а затем обучение самой модели.
model_lr = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

model_ridge = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=1.0))
])

model_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=200, random_state=42))
])

# Обучаем модели на всех данных.
model_lr.fit(X, y)
model_ridge.fit(X, y)
model_rf.fit(X, y)

# Получаем предсказания каждой модели.
pred_lr = model_lr.predict(X)
pred_ridge = model_ridge.predict(X)
pred_rf = model_rf.predict(X)



# Оцениваем качество работы моделей по нескольким метрикам:
# MAE — средняя абсолютная ошибка,
# RMSE — средняя квадратичная ошибка,
# R² — коэффициент детерминации (чем ближе к 1, тем лучше).

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate(true, pred, name):
    print(f"\n===== {name} =====")
    print("MAE :", mean_absolute_error(true, pred))
    print("RMSE:", np.sqrt(mean_squared_error(true, pred)))
    print("R²  :", r2_score(true, pred))

# Выводим метрики для всех трёх моделей.
evaluate(y, pred_lr, "Linear Regression")
evaluate(y, pred_ridge, "Ridge Regression")
evaluate(y, pred_rf, "RandomForest")



# Анализируем важность признаков в модели RandomForest.
# Эта модель может показать, какой признак сильнее всего влияет на прогноз.
rf_model = model_rf.named_steps['regressor']
importances = rf_model.feature_importances_

# Строим график важности признаков.
plt.figure(figsize=(12,4))
plt.bar(range(len(importances)), importances)
plt.title("Feature Importances (RandomForest)")
plt.show()



# Финальный вывод по проекту.
print("\nОБЩИЕ ВЫВОДЫ:")
print("1. Лучшая модель по метрикам — RandomForest.")
print("2. Число голосов (Votes) оказалось самым важным фактором рейтинга.")
print("3. Жанр и длительность фильма также влияют на итоговую оценку IMDb.")
```
